```python
# scales 
import seaborn as sns

for col in [c for c in df.columns if c.startswith("scale_")]:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

### Relationship between two variables 

```python
pd.crosstab(df.jobfield, df.psychotype).plot(kind='bar', figsize=(10,6))
plt.title('Job Field vs Psychotype')
plt.xlabel('Job Field')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Psychotype', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
top_fields = df.jobfield.value_counts().head(10).index 
df_top = df[df.jobfield.isin(top_fields)]

plt.figure(figsize=(10,6))
sns.countplot(data=df_top, x='jobfield', hue='psychotype', palette='Set2')
plt.title('Psychotype Distribution in Top 10 Job Fields')
plt.xlabel('Job Field')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Psychotype', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
crosstab = pd.crosstab(df.jobfield, df.psychotype)
crosstab_norm = crosstab.div(crosstab.sum(axis=1), axis=0).plot(kind='bar', stacked=True, figsize=(10,6), colormap='viridis')
plt.title('Proportional Psychotype Distribution by Job Field')
plt.xlabel('Job Field')
plt.ylabel('Proportion')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Psychotype', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='psychotype', y='scale_e', palette='Set3')
plt.title('Extroversion Scale by Psychotype')
plt.xlabel('Psychotype')
plt.ylabel('Scale E')
plt.show()  
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='psychotype', y='scale_i', palette='Set3')
plt.title('Introversion Scale by Psychotype')
plt.xlabel('Psychotype')
plt.ylabel('Scale IE')
plt.show()
```
