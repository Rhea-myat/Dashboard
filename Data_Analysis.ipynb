```python
%matplotlib inline
import matplotlib.pyplot as plt 
import seaborn as sns
df['psychotype'].hist(bins=30)
plt.show()
```

```python
df['psychotype'].value_counts().plot(kind='bar')
plt.xlabel('Psychotype')
plt.ylabel('Count')
plt.title('Distribution of Psychotypes')
plt.show()
```

```python
sns.catplot(x='psychotype', aspect=1.5, kind='count', data=df)
```

```python
sns.scatterplot(x='psychotype', y='satisfied', data=df)
```

```python
sns.histplot(data=df, x='psychotype', hue='jobfield', multiple='stack', stat='count', shrink=0.8)
plt.xlabel('Psychotype')
plt.ylabel('Count')
plt.title('Distribution of Psychotypes by Job Field')
plt.xticks(rotation=45)
plt.show()
```

```python
# Drop rows with NaN in jobfield or psychotype before crosstab
filtered_df = df.dropna(subset=['jobfield', 'psychotype'])
pd.crosstab(filtered_df.jobfield, filtered_df.psychotype).plot(kind='bar', figsize=(10,6))
plt.title('Job Field vs Psychotype')
plt.xlabel('Job Field')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Psychotype')
plt.show()
```

```python
crosstab = pd.crosstab(df.jobfield, df.psychotype)
plt.figure(figsize=(10,6))
sns.heatmap(crosstab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Job Field vs Psychotype')
plt.xlabel('Psychotype')
plt.ylabel('Job Field')
plt.show()
```

```python
crosstab = pd.crosstab(df.jobfield, df.psychotype)
plt.figure(figsize=(10,6))
sns.heatmap(crosstab,  cmap='YlGnBu', annot=False)
plt.title('Heatmap of Job Field vs Psychotype')
plt.xlabel('Psychotype')
plt.ylabel('Job Field')
plt.show()
```
